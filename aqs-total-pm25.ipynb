```python
def dif_entries(array):
    dif_names = []
    for i in range(len(array)):
        if array[i] not in dif_names:
            dif_names.append(array[i])
    return dif_names
```

```python
dtype_dict = {'Parameter Name': str, 'Sample Duration': str,
              'Units of Measure': str, 'Date Local': str,
              'Units of Measure': str, 'Method Name': str,
              'Arithmetic Mean': str, 'State Name': str} 

cols = ['Parameter Name', 'Units of Measure', 'Sample Duration', 'Method Name', 
        'Date Local', 'Units of Measure', 'Arithmetic Mean', 'State Name']
dfs = []
total_len = 0
for f in glob.glob('C:\\Users\\jenst\\Capstone\\pm_local\\*.csv'):
    df = pd.read_csv(f, dtype = dtype_dict, usecols = cols)
    dfs.append(df)
    total_len += len(df)
#     print(df.dtypes())
daily = pd.concat(dfs, ignore_index=True)  
print('final length = ', len(daily))
```

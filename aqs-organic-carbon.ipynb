```python
def dif_entries(a):
    dif_names = []
    for i in range(len(a)):
        if a[i] not in dif_names:
            dif_names.append(a[i])
    return dif_names
```

```python
dtype_dict = {'Parameter Name': str, 'Sample Duration': str,
              'Metric Used': str, 'Year': str, 'Method Name': str,
              'Units of Measure': str, #'Observation Count': int, 
              'Arithmetic Mean': str, 'State Name': str} 

cols = ['Parameter Name', 'Sample Duration', 'Metric Used', 'Method Name', #'Observation Count', 
        'Year', 'Units of Measure', 'Arithmetic Mean', 'State Name']
dfs = []
total_len = 0
for f in glob.glob('C:\\Users\\jenst\\Capstone\\aqs-yearly\\*.csv'):
    df = pd.read_csv(f, dtype = dtype_dict, usecols = cols)
    dfs.append(df)
    total_len += len(df)
#     print(df.dtypes())
yearly = pd.concat(dfs, ignore_index=True)  
print('final length = ', len(yearly))
```

```python
dif_entries(list(yearly['Parameter Name']))
```

```python
params_interest = ['Organic Carbon Mass PM2.5 LC'] # 'Black carbon PM2.5 STP', 'Black Carbon  PM2.5 Corrected',  'Black Carbon PM2.5 at 880 nm', 'Carbon dioxide', 'Total Carbon PM2.5 LC TOT']
drop_indices = []
for index, row in yearly.iterrows():
    if row['Arithmetic Mean'] == "WGS84":
        drop_indices.append(index)
    elif row['State Name'] != 'Alaska':
        drop_indices.append(index)
    elif row['Parameter Name'] not in params_interest:
        drop_indices.append(index)
print(len(drop_indices))
yearly = yearly.drop(drop_indices)
```

```python
pd.to_numeric(yearly['Arithmetic Mean'])
```

```python
drops = ['Parameter Name', 'Sample Duration', 'Metric Used', 'Method Name', 'Units of Measure', 'State Name']
yearly.drop(columns=drops, inplace=True)
```
